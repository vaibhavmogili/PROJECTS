```python
import torch
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
img = "img1.jpg"
results = model(img)
results.print()
```

```python
cv2.imshow('Result',np.squeeze(results.render()))
cv2.imwrite('new_img1.jpg',np.squeeze(results.render()))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('0'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import uuid   # Unique identifier
import os
import time
labels = ['awake', 'drowsy']
```

```python
#PATH:-> /raw_data/awake
IMAGES_DIRECTORY = os.path.join('raw_data','awake')
IMAGE_PATH=os.path.join('data','images')
image_files=os.listdir(IMAGES_DIRECTORY)
print(IMAGES_DIRECTORY)
count=0
for image_file in image_files:
    count+=1
    temp_path=os.path.join(IMAGES_DIRECTORY,image_file)
    print(temp_path)
    image=cv2.imread(temp_path)
    cv2.imshow('pic',image)
    cv2.waitKey(0)
    
    new_path=os.path.join(IMAGE_PATH,labels[0]+str(count)+'.jpg')
    cv2.imwrite(new_path,image)
    
cv2.destroyAllWindows()
    
```

```python
#PATH:-> /raw_data/drowsy
IMAGES_DIRECTORY = os.path.join('raw_data','drowsy')
IMAGE_PATH=os.path.join('data','images')
image_files=os.listdir(IMAGES_DIRECTORY)
print(IMAGES_DIRECTORY)
count=0
for image_file in image_files:
    count+=1
    temp_path=os.path.join(IMAGES_DIRECTORY,image_file)
    print(temp_path)
    image=cv2.imread(temp_path)
    cv2.imshow('pic',image)
    cv2.waitKey(0)
    
    new_path=os.path.join(IMAGE_PATH,labels[1]+str(count)+'.jpg')
    cv2.imwrite(new_path,image)
    
cv2.destroyAllWindows()
    
```
