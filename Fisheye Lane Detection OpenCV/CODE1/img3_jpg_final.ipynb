```python
#importing the necessary packages
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Reading and displaying the image
original_image=cv2.imread('img3.jpg')
image=cv2.imread('img3.jpg')
height, width = image.shape[:2]

cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Dimesions of the image
height,width,channels=image.shape
print(f"Height = {height}")
print(f"Width = {width}")
print(f"Channels = {channels}")

#Apply (5x5) Gaussian blur to reduce noise
image=cv2.GaussianBlur(image,(5,5),0)
cv2.imshow('blurred image gaussian',image)
cv2.imwrite('Gaussian Blurred image img3.jpg',image)
cv2.waitKey(0)
cv2.destroyAllWindows()

image=image[50:,:]#cropping the image
cv2.imshow('Cropped',image)
rgb_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
gray_image=cv2.cvtColor(rgb_image,cv2.COLOR_RGB2GRAY)
cv2.imwrite('Cropped image img3.jpg',image)
image2=gray_image
cv2.imwrite('Gray Image img3.jpg',image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Applying thresholding Binary on the original image
ret, thresh = cv2.threshold(image, 120,255, cv2.THRESH_BINARY)
cv2.imshow('Original Image - Binary thresholding',thresh)
cv2.imwrite('RGB Image Binary Thresholding img3.jpg',thresh)
#Applying thresholding binary on the gray_image
ret1, thresh1 = cv2.threshold(image2, 120,255, cv2.THRESH_BINARY)
cv2.imshow('Gray Image - Binary thresholding',thresh1)
cv2.imwrite('Gray Image Binary Thresholding img3.jpg',thresh1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hist=cv2.calcHist([image2],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
#Detecting lines using colour segmentation on gray image

colored_image1=np.zeros(image2.shape)
lower_bound1=np.array([118])
upper_bound1=np.array([200])
color1=[0]
color_mask1=cv2.inRange(image2,lower_bound1,upper_bound1)
colored_image1[color_mask1>0]=color1
image2[color_mask1>0]=color1
cv2.imshow("b&w Color segemnted image",image2)
cv2.imwrite('gray color segmented image img3.jpg',image2)
cv2.imshow('b&w color mask',color_mask1)
cv2.imwrite('gray color mask img3.jpg',color_mask1)
cv2.imshow('b&w colored image',colored_image1)
cv2.imwrite('gray colored image img3.jpg',colored_image1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Detecting lines using colour segmentation on rgb image

colored_image2=np.zeros(image.shape)
lower_bound2=np.array([0,100,100])
upper_bound2=np.array([120,240,240])
color2=[0,0,255]
color_mask2=cv2.inRange(image,lower_bound2,upper_bound2)
colored_image2[color_mask2>0]=color2
image[color_mask2>0]=color2
cv2.imshow('rgb color segmented',image)
cv2.imwrite('rgb color segmented image img3.jpg',image)
cv2.imshow('rgb color mask',color_mask2)
cv2.imwrite('rgb color mask img3.jpg',color_mask2)
cv2.imshow('rgb colored image',colored_image2)
cv2.imwrite('rgb colored image img3.jpg',colored_image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
